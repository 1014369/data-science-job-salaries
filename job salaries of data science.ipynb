### Data Science Job Salaries Dataset

### Overview

Data science continues to be one of the fastest-growing fields in technology, with demand and salaries rising globally. The dataset provides salary and job-related information for data professionals across multiple years and locations.

### Feature Description

1.**work_year**: The year the salary was paid.

2.**experience_level**: Job experience level – EN (Entry), MI (Mid), SE (Senior), EX (Executive).

3.**employment_type**: Type of employment – PT (Part-time), FT (Full-time), CT (Contract), FL (Freelance).

4.**job_title**: Specific role held (e.g., Data Scientist, ML Engineer).

5.**salary**: Gross salary paid in local currency.

6.**salary_currency**: Currency of the salary (ISO 4217 format).

7.**salary_in_usd**: Salary converted to USD for standard comparison.

8.**employee_residence**: Country of the employee during the work year (ISO 3166 code).

9.**remote_ratio**: Percentage of remote work – 0 (None), 50 (Partially), 100 (Fully remote).

10.**company_location**: Employer’s main office location.

11.**company_size**: Company size – S (<50), M (50–250), L (>250 employees).
